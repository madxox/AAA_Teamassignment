```python
import pandas as pd
import numpy as np
```

Prepare Weather data

```python
df_weather = pd.read_csv("data/weather_data_2022.csv")
```

```python
display(df_weather)
```

```python
#create timestamp column and name it similar to the one of df_taxi in order to merge the two dataframes based on this column
df_weather['trip_start_timestamp'] = pd.to_datetime(df_weather['Date'] + ' ' + df_weather['Time'])
```

```python
display(df_weather.dtypes)
```

```python
df_weather = df_weather.sort_values('trip_start_timestamp')
```

Prepare Taxi data

(include 'taxidata.csv' to data folder locally)

```python
df_taxi = pd.read_csv("data/prepped/prep_taxidata.csv")
```

```python
display(df_taxi)
```

```python
df_taxi['trip_start_timestamp'] = pd.to_datetime(df_taxi['trip_start_timestamp'])
```

```python
display(df_taxi.dtypes)
```

```python
df_taxi = df_taxi.sort_values('trip_start_timestamp')
```

Merge both data frames based on 'trip_start_timestamp'

```python
df_merged = pd.merge_asof(df_taxi, df_weather, on='trip_start_timestamp', direction='nearest')
```

```python
display(df_merged)
```

```python
print(df_merged.isna().sum())
```

```python
print(df_weather.isna().sum())
```

For 3775 entries in the merged df the 'Wind', which is the wind direction, is missing. This results from only 6 missing values in the original df_weather (because one df_weather entry can be matched to multiple df_taxi entries). I suggest dropping the 'Wind' column, since the wind direction will not be relevant for our purposes.

```python
## We will drop the wind column in future dfs
df_merged.to_csv("data/prepped/weather_taxi_merged_df.csv")
```
